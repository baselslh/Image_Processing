### Choose an image with low contrast and work with it as gray image

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('pout.jpg',0)
```

### Plotting the histogram

```python
plt.hist(img.ravel(), bins=256)
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.show()
```

### Plot the cumulative histogram

```python
plt.hist(img.ravel(), bins=256, cumulative = True)
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.show()
```

### Perform histogram equalization and Plotting the modified image

```python
imgEqualized = cv2.equalizeHist(img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original image')
plt.axis('off')
plt.show()

plt.imshow(cv2.cvtColor(imgEqualized, cv2.COLOR_BGR2RGB))
plt.title('Equalized image')
plt.axis('off')
plt.show()
```

### Plot the modified image histogram

```python
plt.hist(imgEqualized.ravel(), bins=256)
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.show()
```

### Plot the modified image cumulative histogram

```python
plt.hist(imgEqualized.ravel(), bins=256, cumulative = True)
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.show()
```
